```python
class Set :
    def __init__(self, elem = []) :
        
        self.elem = elem
        
        b = []
        self.b = b
        
        for i in self.elem :
            if i not in self.b :
                self.b.append(i)
            else :
                self.b = self.b
                
        
    def add(self, elem) :
        if elem not in self.b :
            self.b.append(elem)
    
    def discard(self, elem) :
        if elem in self.b :
            self.b.remove(elem)
        
    def clear(self) :
        self.b.clear()
        
    def __len__(self) :
        return len(self.b)
    
    def __str__(self) :
        c = str(self.b)
        c = c.replace('[','{',1)
        c = c.replace(']','}',1)
        return c
    
    def __contains__(self, elem) :
        for j in self.b :
            if j == elem :
                return True
            else :
                return False
            
    def __le__(self, other) :
        d = 0
        for k in self.b :
            if k not in other.b :
                d += 1
            else : 
                d += 0
                
        if d == 0 :
            return True
        else :
            return False
        
    def  __ge__(self,other) :
        e = 0
        for l in other.b :
            if l not in self.b :
                e += 1
            else :
                e += 0
        
        if e == 0 :
            return True
        else :
            return False        
    
    def __or__(self,other) :
        import copy
    
        ac = copy.deepcopy(self.b)
        oc = copy.deepcopy(other.b)

        for m in oc :
            if m not in ac :
                ac.append(m)
        
        return Set(ac)
            
        
    def __and__(self, other) :
        
        import copy
        
        ac = copy.deepcopy(self.b)
        oc = copy.deepcopy(other.b)
        
        for o in ac[:] :
            if o not in oc :
                ac.remove(o)
        
        return Set(ac)
            
    def __sub__(self, other) :
        
        import copy
        
        ac = copy.deepcopy(self.b)
        oc = copy.deepcopy(other.b)        
        
        for p in oc :
            if p in ac :
                ac.remove(p)
        
        return Set(ac)
    
    def __ior__(self, other) :
        for q in other.b :
            if q not in self.b :
                self.b.append(q)

        return self
    
    def __iand__(self, other) :
        for o in self.b[:] :
            if o not in other.b :
                self.b.remove(o)
        
        return self
    
    def __isub__(self, other) :
        for p in other.b :
            if p in self.b :
                self.b.remove(p)
        
        return self
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()
```
